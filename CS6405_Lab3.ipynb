```python
import pandas as pd

dataset = pd.read_csv("https://github.com/eleanorformosa/CS6405_Datasets/blob/b662095853ed4bf3e26c9d3a6f7b9a0179e6d3ef/laptop_prices.csv?raw=true")
dataset
```

```python
df = pd.DataFrame(dataset)
df.dtypes
```

```python
df.dtypes
df['Resolution'].unique()
```

```python
objs = pd.get_dummies(df[['Operating System', 'Resolution']], prefix=None, prefix_sep="_", dummy_na=False, columns=None, sparse=False, drop_first=False, dtype=None)
objs
```

```python
merged = pd.concat([df, objs], axis = 1, join='outer', ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=False, copy=True)
```

```python
merged = merged.drop(['GPU', 'Storage', 'Brand', 'Processor'], axis='columns')
```

```python
from sklearn import preprocessing
```

```python
scalingObj = preprocessing.MinMaxScaler()
```

```python
merged
```

```python
new_df = scalingObj.fit_transform(merged[['Battery Life (hours)']])
print(new_df)
```
